```python

import time
import copy
GROUPE07={
    "name":"gestion_des_entre_et_sortie",
    "input": [
        {
            "peripherique": {
               "type":"input",
               "nom":"clavier_standard PS/2",
               "MAC Adress":"A1BC-BE22",
               "status":True,
               "resultat_retourner":["BAC_1"],
            },
            "appel_systeme": {
                "type": "open",
                "params": {
                    "adresse_mem": "0x01A2B3C4",
                    "taille": "512 octets",
                    "buffer": "0x7F4E6B12", 
                    "file_descriptor": 3,
                    "secteur": 1024,
                    "mode_access": "lecture seule"
                },
                "reponse": {
                "status": "succes",
                "donnees": "data lues du clavier_standard PS/2"
                }
            },
            "peripherique": {
               "type":"Input_and_output",
               "nom": Microphone_Array_(Realtek(R)_Audio),
               "MAC Adress":"A1BC-BE22",
               "status":True, False,
               "resultat_retourner":["BAC_1",[1]],
            },
            "appel_systeme": {
                "type": "open",
                "params": {
                    "adresse_mem": "0x01A2B3C4",
                    "taille": "512 octets",
                    "buffer": "0x7F4E6B12", 
                    "file_descriptor": 3,
                    "secteur": 1024,
                    "mode_access": "lecture seule"
                },
                "reponse": {
                "status": "succes",
                "donnees": "data lues du Microphone_Array_"
                }
            },
              "peripherique": {
                "type": "Disque Dur",
                "action": "lecture",
                "params": {
                "adresse_mem": "0x01A2B3C4",
                "taille": "512 octets"
                }
            },
              "appel_systeme": {
                    "type": "read",
                    "params": {
                        "adresse_mem": "0x01A2B3C4",
                        "taille": "512 octets",
                        "buffer": "0x7F4E6B12", 
                        "file_descriptor": 3,
                        "secteur": 1024,
                        "mode_access": "lecture seule"
                    },
                    "reponse": {
                    "status": "succes",
                    "donnees": "data lues du disque"
                    }
                },
        }
    ]
    "Output":[
        {
             "peripherique": {
               "type":"output",
               "nom":"Ecran",
               "MAC Adress":"A1BC-BE22",
               "status":False,
               "resultat_retourner":[1],
            },
            "appel_systeme": {
                "type": "close",
                "params": {
                    "adresse_mem": "0x01A2B3C4",
                    "taille": "512 octets",
                    "buffer": "0x7F4E6B12", 
                    "file_descriptor": 3,
                    "secteur": 1024,
                    "mode_access": "Affichage"
                },
                "reponse": {
                "status": "succes",
                "donnees": "data lecture sur  l'Ecran"
                }
            },
        }
    ]
}


OS={
    "name":"os_bac1_2025",
    "version":"v1.0",
    "modules":[
        GROUPE07
    ],
}


##programmes
import random

# Définition d'un périphérique
class Peripherique:
    def __init__(self, nom, temps_prevu_s):
        self.nom = nom
        self.temps_prevu_s = temps_prevu_s  # Temps prévu pour l'exécution du périphérique
        self.temps_ecoule_s = 0  # Temps écoulé pour l'exécution du périphérique

    def executer(self):
        """
        Simule l'exécution d'un périphérique.
        On incrémente le temps écoulé et on vérifie si le périphérique est terminé.
        """
        # Si le périphérique n'a pas encore terminé
        if self.temps_ecoule_s < self.temps_prevu_s:
            # Incrémenter le temps écoulé
            time.sleep(1)  # Simuler un délai de 1 seconde pour chaque étape d'exécution
            self.temps_ecoule_s += 1
            
            print(f"Temps écoulé sur {self.nom}: {self.temps_ecoule_s}/{self.temps_prevu_s} secondes")
            return True  # Le périphérique n'est pas encore terminé
        
        else:
            # Si le temps écoulé atteint ou dépasse le temps prévu
            print(f"{self.nom} est terminé.")
            return False  # Le périphérique est terminé

# Exemple d'exécution d'un périphérique
def gestion_peripherique():
    # Initialiser un périphérique (par exemple un disque dur)
    disque_dur = Peripherique("Disque Dur", temps_prevu_s=5)
    carte_reseau = Peripherique("Carte Réseau", temps_prevu_s=3)

    # Liste des périphériques à exécuter
    peripheriques = [disque_dur, carte_reseau]

    # Exécution des périphériques
    while True:
        tous_termines = True
        for peripherique in peripheriques:
            if peripherique.executer():
                tous_termines = False
        if tous_termines:
            print("Tous les périphériques ont terminé leur exécution.")
            break  # Sortir de la boucle une fois que tous les périphériques sont terminés

# Appel de la fonction pour exécuter les périphériques
gestion_peripherique()
```
